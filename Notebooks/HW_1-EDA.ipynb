## 2

Посчитать статистики: 0.05 и 0.95 квантили, медиана, средняя, дисперсия для любой из указанных величин на выбор: размер трейда, объём какого-то фиксированного уровня стакана

```python
def get_statistics(array):
    # Quantile 0.05
    q_005 = np.quantile(array, 0.05)

    # Quantile 0.95
    q_095 = np.quantile(array, 0.95)

    # Median
    med = array.median()

    # Mean
    mean = array.mean()

    # Dispersion
    disp = array.std()

    return {
        'quantile_0.05': q_005,
        'quantile_0.95': q_095,
        'median': med,
        'mean': mean,
        'dispersion': disp
    }
```

```python
print('BTC-size\n', get_statistics(btc_trades['size']), end='\n\n')
print('BTC-size(USD)\n', get_statistics(btc_trades['size']*btc_trades['price']), end='\n\n')
# логика в том, что цена битка больше цены эфира

print()

print('ETH-size\n', get_statistics(eth_trades['size']), end='\n\n')
print('ETH-size(USD)\n', get_statistics(eth_trades['size']*eth_trades['price']), end='\n\n')
```

У битка все в 2 раза больше (чем у эфира) относительно (USD). Возможно большая популярность дает более крупных игроков.

Как-то так.

## 4

Также нужно построить график величины "пробития" стакана от размера трейда того, насколько "пробивается" стакан в результате трейда. Величиной пробития стакана называется величина изменения best_price в результате трейда. Считать также стоит раздельно для каждой стороны.

Под пробитием стакана я понимаю "ретроспективное пробитие". Т.е. как и когда пробивался стакан в ходе торгов, а не на сколько в какой-то момент времени можно пробить стакан (в этом случае график совпадает с графиком ask/bid_cumvol_ )

```python
btc_lobs
```

```python
btc_trades
```

```python
plt.plot(btc_lobs['ask_cumvol_0'])
```

```python
plt.plot(btc_trades['size'])
```

```python
for i in range(10):
    d = dict(zip(btc_lobs[' exchange_ts'], btc_lobs['ask_cumvol_' + str(i)]))
    btc_trades['ask_cumvol_' + str(i)] = btc_trades['exchange_ts'].map(d)

    d = dict(zip(btc_lobs[' exchange_ts'], btc_lobs['bid_cumvol_' + str(i)]))
    btc_trades['bid_cumvol_' + str(i)] = btc_trades['exchange_ts'].map(d)

    d = dict(zip(btc_lobs[' exchange_ts'], btc_lobs['btcusdt:Binance:LinearPerpetual_ask_price_' + str(i)]))
    btc_trades['ask_price_' + str(i)] = btc_trades['exchange_ts'].map(d)

    d = dict(zip(btc_lobs[' exchange_ts'], btc_lobs['btcusdt:Binance:LinearPerpetual_bid_price_' + str(i)]))
    btc_trades['bid_price_' + str(i)] = btc_trades['exchange_ts'].map(d)
```

```python
btc_trades_rr = btc_trades.dropna()
btc_trades_rr
```

```python
shift = []
shift_usd = []

# посчитаем насколько пробивался стакан и насколько переплачивали люди

for index, row in btc_trades_rr.iterrows(): # извиняюсь за циклы
    ans = 0
    ans_usd = 0
    if row['aggro_side'] == 'ASK':
        for i in range(10):
            if row['ask_cumvol_' + str(i)] < row['size']:
                ans += 1
        ans_usd = row['ask_price_' + str(min(ans, 4))] - row['ask_price_0']
        shift.append(ans)
        shift_usd.append(ans_usd)
    else:
        for i in range(10):
            if row['ask_cumvol_' + str(i)] < row['size']:
                ans += 1
        ans_usd = row['ask_price_' + str(min(ans, 4))] - row['ask_price_0']
        shift.append(-ans)
        shift_usd.append(-ans_usd)

btc_trades_rr['shift'] = shift
btc_trades_rr['shift_usd'] = shift_usd
```

### Теперь подумаем про отставание в объемах торгов

```python
def mae(a, b):
    ab = pd.DataFrame({
        'a': a,
        'b': b
    })
    ab = ab.dropna()
    a = ab['a']
    b = ab['b']

    return abs(np.array(a) - np.array(b)).mean()
```

```python
btc_trades_dim_size = btc_trades['size'] * (2*(btc_trades['aggro_side'] == 'ASK') - 1)
eth_trades_dim_size = eth_trades['size'] * (2*(eth_trades['aggro_side'] == 'ASK') - 1)
```
